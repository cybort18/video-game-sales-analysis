```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vgsales.csv')
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
genre_sales = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
genre_sales
```

```python
df_clean = df.dropna()
print(df_clean.isnull().sum())
```

```python
regional_sales = df_clean.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales']].sum()

regional_sales.plot(kind='bar', figsize=(12, 6))
plt.title('Perbandingan Penjualan Video Game per Wilayah Berdasarkan Genre')
plt.xlabel('Genre')
plt.ylabel('Total Penjualan (Juta Kopi)')
plt.xticks(rotation=45)
plt.legend(['Amerika Utara (NA)', 'Eropa (EU)', 'Jepang (JP)'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
top_publishers = df_clean.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)

top_publishers.plot(kind='bar', figsize=(10, 5), color='coral')
plt.title('Top 10 Publisher Video Game Berdasarkan Penjualan Global')
plt.xlabel('Publisher')
plt.ylabel('Total Global Sales (Juta Kopi)')
plt.xticks(rotation=45)
plt.show()
```

```python
top_platforms = df_clean.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10)

top_platforms.plot(kind='bar', figsize=(10, 5), color='mediumseagreen')
plt.title('Top 10 Platform Video Game Berdasarkan Penjualan Global')
plt.xlabel('Platform')
plt.ylabel('Total Global Sales (Juta Kopi)')
plt.xticks(rotation=0)
plt.show()
```

```python
genre_sales.plot(kind='bar')
plt.title('Total Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.show()
```
